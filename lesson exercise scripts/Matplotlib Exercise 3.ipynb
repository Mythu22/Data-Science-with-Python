```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
```

```python
fig, (ax1, ax2)  = plt.subplots(1, 2, sharey = True)
ax1.plot(x, y)
ax2.scatter(x, y)
```

```python
fig, axes = plt.subplots(2, 2, figsize = (15, 15))
axes[0, 0].plot(x, y)
axes[0, 1].plot(x**2, y)
axes[1, 0].scatter(x**2, y**2)
axes[1, 1].scatter(x, y)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.random.rand(20)
y = np.random.rand(20)
z = np.sqrt(x**2 + y**2)

plt.figure(figsize = (15, 20))
plt.subplot(321)
plt.scatter(x, y, s=80, c=z, marker=">")

plt.subplot(322)
plt.scatter(x, y, s=80, c=z, marker='*')

plt.subplot(323)
plt.scatter(x, y, s=80, c=z, marker= 'o')
# equivalent:
# plt.scatter(x, y, s=80, c=z, marker=None, verts=verts)

plt.subplot(324)
plt.scatter(x, y, s=80, c=z, marker=',')

plt.subplot(325)
plt.scatter(x, y, s=80, c=z, marker='+')

plt.subplot(326)
plt.scatter(x, y, s=80, c=z, marker= 'p')

plt.show()
```
