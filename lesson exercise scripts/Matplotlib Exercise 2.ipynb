### Scatter Plot

```python
import numpy as np
import matplotlib.pyplot as plt

n = 50
x = np.random.randn(n)
y = np.random.randn(n)

colors = np.random.randn(n)
size = (20*np.random.randn(n))**2

plt.scatter(x, y, s = size, c = colors, cmap = 'summer', alpha = 0.5)
plt.show()
```
