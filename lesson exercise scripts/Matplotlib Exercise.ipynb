```python
# Import Pyplot Module
import matplotlib.pyplot as plt

# Plot
x = [1, 2, 3]
y = [3, 4, 5]

plt.plot(x, y, 'go--', linewidth = 2, markersize = 12)

plt.title(' Plot ')
plt.ylabel(' Y axis')
plt.xlabel(' X axis ')

# Show Plot
plt.show()
```

```python
import numpy as np

z = np.arange(1, 10, 0.5)

plt.plot(z, z, 'r--', z, z**2, 'bs', z, z**3, 'g+')
plt.show()
```
